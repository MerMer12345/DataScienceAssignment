Obesity time/Total/all states

```python
import pandas as pd
import matplotlib.pyplot as plt

df_Obesity = pd.read_csv('Data/Obesity/Obesity_file_total.csv')
#df_show = df_Obesity[['YearStart', 'Data_Value']]

plt.style.use('ggplot')

plt.plot(df_Obesity['YearStart'], df_Obesity['Data_Value'])
plt.xlabel('Year')
plt.ylabel('% Of people with Obesity')

```

```python

temp = df_Obesity[df_Obesity['LocationAbbr'] == 'AK']
df_show = temp[['YearStart', 'Data_Value']]
plt.style.use('ggplot')
plt.xlabel('Year')
plt.ylabel('Percentage of Obesity in Alaska')
plt.plot(df_show['YearStart'], df_show['Data_Value'])

plt.show()
```

```python
df2 = pd.read_csv('Data/Obesity/Obesity_file_Education.csv')
plt.figure(figsize=(10,6))
plt.xlabel('Education')
plt.ylabel('Percentage of Obesity')
plt.bar(df2['Education'], df2['Data_Value'], width=0.5)

plt.show();
```

```python
df2 = pd.read_csv('Data/Obesity/Obesity_file_Income.csv')
df2 = df2.sort_values(by=['Income'])
plt.figure(figsize=(13,6)) 
plt.xlabel('Income')
plt.ylabel('Percentage of Obesity')
plt.bar(df2['Income'], df2['Data_Value'], width=0.5)
plt.show();
```

```python
df2 = pd.read_csv('Data/Obesity/Obesity_file_Gender.csv')
plt.xlabel('Gender')
plt.ylabel('Percentage of Obesity')
plt.bar(df2['Gender'], df2['Data_Value'], width=0.5)  
plt.show();
```
